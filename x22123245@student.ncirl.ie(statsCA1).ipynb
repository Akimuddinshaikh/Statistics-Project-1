```python
#calculating mortality rate
df['mortalityRate'] = df['deathRate'] / df['incidenceRate']
```

# Plotting Mortality rate with just incidence rate 

```python
import statsmodels.api as sm
X_1= df[["incidenceRate"]]
y_1 = df['mortalityRate']
# Fit the multiple linear regression model
model_1 = sm.OLS(y_1, X_1).fit()

# Print the model summary
print(model_1.summary())

# Plot the residuals vs. fitted values
residuals_1 = model_1.resid
fitted_1 = model_1.fittedvalues
plt.scatter(fitted_1, residuals_1)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot for the independent varible as Incidence rate only ')
plt.show()
```

# Second Model 

```python
import pandas as pd
import statsmodels.api as sm


# Derive the mortality rate

# Select the predictor variables
X = df[['incidenceRate','medIncome', 'PctUnemployed16_Over', 'povertyPercent', 'MedianAge', 'AvgHouseholdSize', 'PctMarriedHouseholds', 'PctHS18_24', 'PctBachDeg25_Over',]]

# Add a constant term to the predictor variables
X = sm.add_constant(X)

# Define the response variable
y = df['mortalityRate']

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

# Plot the residuals vs. fitted values
residuals = model.resid
fitted = model.fittedvalues
plt.scatter(fitted, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

# Gauss Markov check 

```python

# Linearity test
fig, axs = plt.subplots(1, len(X.columns), figsize=(20,5))
for i, col in enumerate(X.columns):
    axs[i].scatter(X[col], y)
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Death Rate')
plt.show()
```

```python
# Normality test
residuals = model.resid
fig, axs = plt.subplots(1, 2, figsize=(10,5))
sm.graphics.qqplot(residuals, line='r', ax=axs[0])
axs[0].set_title('Normal Probability Plot')
axs[1].hist(residuals)
axs[1].set_title('Histogram of Residuals')
plt.show()
```

```python

# Homoscedasticity test
plt.scatter(model.fittedvalues, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
# Multicollinearity test
import numpy as np
corr_matrix = np.corrcoef(X, rowvar=False)
fig, ax = plt.subplots()
cax = ax.imshow(corr_matrix, interpolation='nearest', cmap='coolwarm')
ax.set_xticklabels(X.columns)
ax.set_yticklabels(X.columns)
plt.colorbar(cax)
plt.show()
```

```python
# Outlier detection
fig, ax = plt.subplots()
ax.scatter(model.fittedvalues, residuals)
ax.set_xlabel('Predicted Values')
ax.set_ylabel('Residuals')
ax.set_title('Residuals vs. Predicted Values')
ax.axhline(y=0, color='r', linestyle='-')
for i, txt in enumerate(X.index):
    if residuals[i] > 2*np.std(residuals) or residuals[i] < -2*np.std(residuals):
        ax.annotate(txt, (model.fittedvalues[i], residuals[i]))
plt.show()
```

```python
# Influence detection
influence = model.get_influence()
(c, _) = influence.cooks_distance
fig, ax = plt.subplots(figsize=(12,6))
ax.stem(np.arange(len(X)), c, markerfmt=",")
ax.set_xlabel("Row index")
ax.set_ylabel("Cook's distance")
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["feature"] = X.columns
print(vif)
```

# Models for the all variables 

```python
X_all = df[['Population', 'incidenceRate', 'medIncome',
       'povertyPercent', 'MedianAge', 'MedianAgeMale', 'MedianAgeFemale',
       'AvgHouseholdSize', 'PctMarriedHouseholds', 'PctNoHS18_24',
       'PctHS18_24', 'PctBachDeg18_24', 'PctHS25_Over', 'PctBachDeg25_Over',
       'PctUnemployed16_Over', 'PctPrivateCoverage', 'PctEmpPrivCoverage',
       'PctPublicCoverage', 'PctPublicCoverageAlone', 'PctWhite', 'PctBlack',
       'PctAsian', 'PctOtherRace']]
y_all = df["mortalityRate"]
```

```python
model_all = sm.OLS(y_all, X_all).fit()

# Print the model summary
print(model_all.summary())

# Plot the residuals vs. fitted values
residuals = model_all.resid
fitted = model_all.fittedvalues
plt.scatter(fitted, residuals)
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot taking all the variables ')
plt.show()
```

```python
# Normality test
residuals = model_all.resid
fig, axs = plt.subplots(1, 2, figsize=(10,5))
sm.graphics.qqplot(residuals, line='r', ax=axs[0])
axs[0].set_title('Normal Probability Plot')
axs[1].hist(residuals)
axs[1].set_title('Histogram of Residuals')
plt.show()
```

```python
# Homoscedasticity test
plt.scatter(model_all.fittedvalues, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```
